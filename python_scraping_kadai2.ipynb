```python
import requests
from getpass import getpass

# 検索条件を入力
api_key = getpass('APIキーを入力してください: ')
search_address = input('検索する駅を入力してください: ')

# geocodeから検索駅のjsondataを取得
geocode_url = 'https://maps.googleapis.com/maps/api/geocode/json'
params = {
    'address': search_address,
    'key': api_key,
}
response = requests.get(geocode_url, params=params)
search_json = response.json()

# jsondataから緯度経度を取得
if search_json['status'] == 'OK':
    location = search_json['results'][0]['geometry']['location']
    latitude = location['lat']
    longitude = location['lng']
    print(f"{search_address}の情報を取得しました。")

    # 検索条件を入力
    keyword = input(f'{search_address}のどんなお店をお探しですか: ')
    area_value = input(f'{search_address}から何メートル以内を検索しますか: ')

    # 検索駅から条件を基に店舗のjsondataを取得
    store_url = f'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={latitude},{longitude}&radius={area_value}&type=restaurant&keyword={keyword}&key={api_key}'

    response = requests.get(store_url,params = {"language": "ja"})
    store_json = response.json()

    # jsondataから店舗情報を出力
    if 'results' in store_json:
        stores = store_json['results']

        if 'status' in store_json and store_json['status'] == 'ZERO_RESULTS':
            print(f'{search_address}から{area_value}m以内にある{keyword}の情報が見つかりませんでした。')
        else:
            # 'status' フィールドが存在しないか、'ZERO_RESULTS' 以外の値の場合
            for store in stores:
                name = store['name']
                rating = store.get('rating', '評価なし')
                address = store.get('vicinity', '住所なし')
                print(f'店舗名: {name}')
                print(f'評価: {rating}')
                print(f'住所: {address}')
                print('-' * 50)
    else:
        print(f'{keyword}の情報を正しく取得できませんでした')
else:
    print(f"{search_address}の情報を取得できませんでした。")
```
